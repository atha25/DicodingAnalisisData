```python
Q1 = df_bikeday["cnt"].quantile(0.25)
Q3 = df_bikeday["cnt"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_bikeday[(df_bikeday["cnt"] < lower_bound) | (df_bikeday["cnt"] > upper_bound)]
print("Outliers in 'cnt' column of df_bikeday:")
display(outliers.head().style.background_gradient(cmap="viridis"))

plt.figure(figsize=(8, 6))
sns.boxplot(x=df_bikeday['cnt'])
plt.title('Box Plot of Bike Rentals per Hari (df_bikeday)')
plt.xlabel('Count (cnt)')
plt.show()
```

```python
Q1 = df_bikehour["cnt"].quantile(0.25)
Q3 = df_bikehour["cnt"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_bikehour[(df_bikehour["cnt"] < lower_bound) | (df_bikehour["cnt"] > upper_bound)]
print("Outliers in 'cnt' column of df_bikehour:")
display(outliers.head().style.background_gradient(cmap="viridis"))

plt.figure(figsize=(8, 6))
sns.boxplot(x=df_bikehour['cnt'])
plt.title('Box Plot of Bike Rentals per Jam (df_bikehour)')
plt.xlabel('Count (cnt)')
plt.show()
```

**Insight:**
- File Bike Day tidak mempunyai Outlier
- File Bike Hour punya Outlier
- Dataset tidak punya Null/NaN value
- Dataset tidak punya Duplicate value
- Nilai Outlier dalam hour.csv dapat mengindikasikan jam-jam tertentu dengan lonjakan atau penurunan sewa yang luar biasa tinggi, seperti selama jam sibuk atau kondisi cuaca ekstrem.
- Tidak adanya outlier dalam day.csv menunjukkan bahwa tren sewa harian relatif stabil, tanpa fluktuasi ekstrem yang signifikan.

### Cleaning Data

```python
df_bikeday_cleaned_iqr = df_bikeday[(df_bikeday["cnt"] >= lower_bound) & (df_bikeday["cnt"] <= upper_bound)].copy()

print(f"Bentuk original df_bikeday: {df_bikeday.shape}")
print(f"Bentuk df_bikeday setelah menghapus outliers: {df_bikeday_cleaned_iqr.shape}")

display(df_bikeday_cleaned_iqr.head())
```

```python
df_bikehour_cleaned_iqr = df_bikehour[(df_bikehour["cnt"] >= lower_bound) & (df_bikehour["cnt"] <= upper_bound)].copy()

print(f"Bentuk original df_bikehour: {df_bikehour.shape}")
print(f"Bentuk df_bikehour setelah menghapus outliers: {df_bikehour_cleaned_iqr.shape}")

display(df_bikehour_cleaned_iqr.head())
```

**Insight:**
- Outlier terdeteksi pada dataset hour.csv dan telah ditangani menggunakan metode IQR untuk mengurangi pengaruh nilai ekstrem yang dapat mengganggu analisis.

- Dataset day.csv tidak memiliki outlier signifikan, menunjukkan distribusi data harian yang lebih konsisten.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure()
plt.plot(df_bikeday['dteday'], df_bikeday['casual'], label='Casual', color='blue')
plt.plot(df_bikeday['dteday'], df_bikeday['registered'], label='Registered', color='orange')

plt.title('Daily Bike Rentals: Casual vs Registered')
plt.xlabel('Date')
plt.ylabel('Number of Rentals')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

- Rata-rata penyewaan relatif tinggi dan stabil sepanjang hari kerja, dengan sedikit peningkatan pada pertengahan hingga akhir minggu, menandakan penggunaan konsisten untuk rutinitas harian.
- Musim Fall (musim gugur) memiliki jumlah penyewaan tertinggi, diikuti Summer dan Winter, sedangkan Spring menunjukkan penggunaan paling rendah, mencerminkan pengaruh kuat kondisi cuaca terhadap permintaan.

## Analisis Lanjutan (Opsional)

```python
def time_bin(hr):
    if hr <= 5:
        return "Subuh"
    elif hr <= 10:
        return "Pagi"
    elif hr <= 15:
        return "Siang"
    elif hr <= 20:
        return "Sore/Malam"
    else:
        return "Tengah Malam"

df_bikehour['Time_Period'] = df_bikehour['hr'].apply(time_bin)
df_bikehour.groupby('Time_Period')['cnt'].mean()
```

**Insight**
- Rata-rata penyewaan sepeda tertinggi terjadi pada waktu Sore/Malam, diikuti Siang dan Pagi, menunjukkan bahwa permintaan meningkat pada jam pulang kerja dan aktivitas sore hari.
